### What is Deep Learning ?

*  Deep Learning is a subfield of Artificial Intelligence and Machine Learning that is inspired by the structure of human brain.
* DL algorithms attempt to draw similar conclusions as human would by continuously analyzing data with a given logical structure called Neural Network.
* DL is a part of a broder family of machine learning methods based on artificial neural networks with representation learning(also called feature extraction where ml requires features creation for classifying from dog and cat while DL did it my just analyzing the images so there is no need to create any features in it.)
* DL algorithms uses multiple layers to progressively extract higher level features from the raw input.for example, in image processing,lower layers may identify edges,while higher layers may identify the concepts relevant to a human such as digits or letters or faces.

#### Deep Learning V/S Machine Learning
(DL is subset of ML as ML algorithms are relay on statistical methods while DL is based on human brains logical structure called Neural Network. )

1. Data Dependency :
  * DL requires more data to learn and give great results than compare to the ML.
  * according to famous curse between them,if data is less then ML performs better than DL but if data is vast then ML start to stagnant the performance while DL increases the performance. hence, DL is data hungry.

2. Hardware Dependency :
 * ML can perform on CPU but DL has complex matrix multiplication which requires powerful GPU , if we use CPU for it then run slow.hence expensive hardware required for DL.

3. Training Time :
* DL is complex so requires more time for training than ML.
* but prediction time varies in ML and DL as in ML KNN requires much time to predict.

4. Feature Selection :
*  in DL there is representational learning so we dont need to extract features while in ML we have to manually extract the features.

5. Interpretability :
* everything is happening internally in DL which we cannot explain the reason behind it., but in ML interpretibility is high
