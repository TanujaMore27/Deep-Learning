### Perceptron Trick :
* We first scatter plot of data points and take a random line separating regions.
* then take one point and check if he is in right region or not if no then pull that line towards the data point such that it will in in its correct region.
* this can be done in 1k or 10k epochs or till the convergence meaning after every line shifting check how many points are misclassified so do it till misclassify points becomes zero.
* hence as points pull or push the line the value of A,B,C changes and the last epoch or convergence value becomes the main values for classifying i.e in DL its weights and bias.

#### How to label the regions (if line becomes / from \) :
* to know how to classify points in region so as per equation AX+BY+C=0 , put that data point value in x,y then if AX+BY+C >= 0 it will be positive region and if AX+BY+C < 0 then it will be classified as the negative region.

```python
from sklearn.datasets import make_classification
import numpy as np
x,y = make_classification(n_samples=100,n_features=2,n_informative=1,n_redundant=0,n_classes=2,n_clusters_per_class=1,random_state=41,hypercube=False,class_sep=10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10,6))
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
```

```python
def perceptron(x,y):
  x = np.insert(x,0,1,axis=1)
  weights = np.ones(x.shape[1])
  lr = 0.1

  for j in range(1000):
    j = np.random.randint(0,100)
    y_hat = step(np.dot(x[j],weights))
    weights = weights + lr*(y[j]-y_hat)*x[j]

  return weights[0],weights[1:]
```

```python
def step(z):
  return 1 if z>0 else 0
```

```python
intercept_,coef_ = perceptron(x,y)
```

```python
print(coef_)
print(intercept_)
```

```python
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
```

```python
x_input = np.linspace(-3,3,100)
y_input = m*x_input + b
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_input,y_input,color='red',linewidth=3)
plt.scatter(x[:,0],x[:,1],c=y,cmap='winter',s=100)
plt.ylim(-3,2)
```
